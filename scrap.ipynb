```python
url=[]
name=[]
price=[]
rating=[]
review=[]

for i in range(21):
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    webpage=requests.get('https://www.amazon.in/s?k=bags&crid=2M096C61O4MLT&qid=1653308124&sprefix=ba%2Caps%2C283&ref=sr_pg_{}'.format(i),headers=headers).text

    soup=BeautifulSoup(webpage,'lxml')
    s=0
    #basket=soup.find_all('div',class_="sg-col-inner")
    basket=soup.find_all('div',class_="a-section a-spacing-small a-spacing-top-small")
    for i in basket[1:]:
        name.append(i.find('span',class_="a-size-medium a-color-base a-text-normal").text)
     

        
        
        try:
            url.append(i.find_all('a',class_="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal",href=True))
            
        except:
            url.append(np.nan)
        try:
            price.append(i.find('span',class_="a-price-whole").text)
        except:
            price.append(np.nan)
        try:
            rating.append(i.find('span',class_="a-icon-alt").text[:3])
        except:
            rating.append(np.nan)
        try:
            review.append(i.find('span',class_="a-size-base s-underline-text").text)
        except:
            review.append(np.nan)


        
        s+=1                   #to keep a count on data recieved
    print(s)    
#for i in soup.find_all('a',class_="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal",href=True):
    #print(i['href'])  '''      
```
